```python
# Import helpful libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# Load the data, and separate the target
file_path = '../input/home-data-for-ml-course/train.csv'
home_data = pd.read_csv(file_path)
```

```python
home_data.head()
```

```python
home_data.info()
```

```python
features = ['MSSubClass','LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold','YrSold','SalePrice']
home_data[features]
```

```python
plt.figure(figsize = (30,30))
plt.grid(axis = 'y', alpha = 0.5)
ax = sns.heatmap(home_data[features].corr(),annot=True,cmap='viridis')
ax.set_title('Pearson Heat Map', fontsize = 20)
```

```python
Y = home_data.SalePrice
good_features = ['LotArea','OverallQual','YearBuilt','YearRemodAdd','BsmtFinSF1','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','GrLivArea','BsmtFullBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','ScreenPorch']
X = home_data[good_features]
X.head()
```

```python
X.info()
```

```python
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X,Y)
```

```python
test_data_path = '../input/home-data-for-ml-course/test.csv'
test_data = pd.read_csv(test_data_path)
test_X = test_data[good_features]
```

```python
test_X.info()
```

```python
test_X['GarageCars'] = test_X['GarageCars'].fillna(test_X['GarageCars'].median())
test_X['BsmtFullBath'] = test_X['BsmtFullBath'].fillna(test_X['BsmtFullBath'].median())
test_X['BsmtFinSF1'] = test_X['BsmtFinSF1'].fillna(test_X['BsmtFinSF1'].mean())
test_X['BsmtUnfSF'] = test_X['BsmtUnfSF'].fillna(test_X['BsmtUnfSF'].mean())
test_X['TotalBsmtSF'] = test_X['TotalBsmtSF'].fillna(test_X['TotalBsmtSF'].mean())
test_X['GarageArea'] = test_X['GarageArea'].fillna(test_X['GarageArea'].mean())
test_X.info()
```

```python
test_preds = rf_model.predict(test_X)
output = pd.DataFrame({'Id': test_data.Id,
                       'SalePrice': test_preds})
output.to_csv('submission2.csv', index=False)
```
